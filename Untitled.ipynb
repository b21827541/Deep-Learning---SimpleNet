```python
class MVTecTestDataset(MVTecDataset):
    """Extended MVTec dataset class that also returns mask and file path."""
    
    def __init__(self, base_path, category, split='test', transform=None):
        super().__init__(base_path, category, split, transform)
    
    def _load_dataset(self):
        """Load dataset with additional mask information."""
        path = self.base_path / self.category / self.split
        
        # Get all image paths
        if self.split == 'train':
            # Training set only contains good images
            image_paths = list(path.glob('good/*.png'))
            labels = [0] * len(image_paths)  # 0 for normal
            masks = [None] * len(image_paths)  # No masks for normal images
        else:
            # Test set contains both good and defective images
            image_paths = []
            labels = []
            masks = []
            
            for defect_type in path.iterdir():
                if defect_type.is_dir():  # Skip any non-directory files
                    curr_paths = list(defect_type.glob('*.png'))
                    image_paths.extend(curr_paths)
                    
                    # 0 for normal (good), 1 for anomaly
                    is_anomaly = 0 if defect_type.name == 'good' else 1
                    curr_labels = [is_anomaly] * len(curr_paths)
                    labels.extend(curr_labels)
                    
                    # Get masks for anomalous images
                    if is_anomaly:
                        # Construct mask paths from image paths
                        curr_masks = []
                        for img_path in curr_paths:
                            # Example: if image is at "category/test/defect/001.png"
                            # mask should be at "category/ground_truth/defect/001_mask.png"
                            mask_dir = self.base_path / self.category / 'ground_truth' / defect_type.name
                            mask_path = mask_dir / f"{img_path.stem}_mask.png"
                            curr_masks.append(mask_path if mask_path.exists() else None)
                        masks.extend(curr_masks)
                    else:
                        # No masks for normal images
                        masks.extend([None] * len(curr_paths))
        
        return list(zip(image_paths, labels, masks))
    
    def __getitem__(self, idx):
        image_path, label, mask_path = self.samples[idx]
        
        with Image.open(image_path) as img:
            img = img.convert('RGB')
            if self.transform:
                img = self.transform(img)
        
        # Load mask if available
        if mask_path is not None and os.path.exists(mask_path):
            with Image.open(mask_path) as mask_img:
                mask = np.array(mask_img) > 0  # Convert to binary mask
                mask = torch.tensor(mask, dtype=torch.float32)
        else:
            mask = torch.zeros((224, 224))  # Empty mask for normal images
        
        return img, label, mask, str(image_path)
```

```python
def get_test_dataloader(base_path, category, batch_size=32):
    """Create test dataloader with mask information."""
    transform = get_transform()
    
    # Create dataset
    test_dataset = MVTecTestDataset(Path(base_path), category, 'test', transform)
    
    # Custom collate function to handle masks of different sizes
    def custom_collate(batch):
        images = torch.stack([item[0] for item in batch])
        labels = torch.tensor([item[1] for item in batch])
        # Don't stack masks because they might have different sizes
        masks = [item[2] for item in batch]
        image_paths = [item[3] for item in batch]
        return images, labels, masks, image_paths
    
    # Create dataloader with custom collate function
    test_loader = DataLoader(
        test_dataset,
        batch_size=batch_size,
        shuffle=False,
        num_workers=4,
        pin_memory=True,
        collate_fn=custom_collate
    )
    
    return test_loader
```

```python
def load_trained_model(checkpoint_path, device):
    """Load trained models from checkpoint."""
    # Initialize models
    feature_extractor, adapter, discriminator = initialize_models(device)
    
    # Load checkpoint
    checkpoint = torch.load(checkpoint_path, map_location=device)
    
    # Remove 'module.' prefix from state_dict keys if it exists (for DDP training)
    adapter_state_dict = checkpoint['adapter_state_dict']
    discriminator_state_dict = checkpoint['discriminator_state_dict']
    
    # Check if state_dict has 'module.' prefix and remove it
    if all(k.startswith('module.') for k in adapter_state_dict.keys()):
        adapter_state_dict = {k[7:]: v for k, v in adapter_state_dict.items()}
        
    if all(k.startswith('module.') for k in discriminator_state_dict.keys()):
        discriminator_state_dict = {k[7:]: v for k, v in discriminator_state_dict.items()}
    
    # Load state dictionaries
    adapter.load_state_dict(adapter_state_dict)
    discriminator.load_state_dict(discriminator_state_dict)
    
    return feature_extractor, adapter, discriminator
```

```python
def generate_anomaly_maps(feature_extractor, adapter, discriminator, test_loader, device, save_dir=None):
    """Generate anomaly maps for test images."""
    image_scores = []
    pixel_scores = []
    true_labels = []
    true_masks = []
    file_names = []
    
    # Set models to evaluation mode
    feature_extractor.eval()
    adapter.eval()
    discriminator.eval()
    
    with torch.no_grad():
        for images, labels, masks, image_paths in tqdm(test_loader, desc="Generating anomaly maps"):
            images = images.to(device)
            
            # Get features
            feats = feature_extractor(images)
            combined_feats = pool_and_combine_features(feats)
            embedding = flatten_features(combined_feats)
            adapted = adapter(embedding)
            
            # Get anomaly scores
            scores = discriminator(adapted)
            # Define anomaly score as -D(q) (paper Eq. 9)
            scores = -scores
            
            B = images.shape[0]
            H, W = combined_feats.shape[2], combined_feats.shape[3]
            
            # Reshape for image-level and pixel-level scores
            scores_map = scores.view(B, H, W)
            
            # Image-level scores (max over spatial dimensions)
            img_scores = scores_map.max(dim=1)[0].max(dim=1)[0]
            
            # For each image in the batch
            for i in range(B):
                # Store results
                anomaly_map = scores_map[i].cpu().numpy()
                
                # Resize to original image size and apply gaussian filter
                anomaly_map = cv2.resize(anomaly_map, (224, 224))
                anomaly_map = gaussian_filter(anomaly_map, sigma=4)
                
                image_scores.append(img_scores[i].cpu().numpy())
                pixel_scores.append(anomaly_map)
                true_labels.append(labels[i].cpu().numpy())
                
                # Handle mask (now a tensor or None, not a batch)
                if masks[i] is not None:
                    mask_np = masks[i].cpu().numpy()
                    if mask_np.shape != anomaly_map.shape:
                        mask_np = cv2.resize(mask_np, (anomaly_map.shape[1], anomaly_map.shape[0]))
                    true_masks.append(mask_np)
                else:
                    true_masks.append(np.zeros_like(anomaly_map))
                
                file_names.append(os.path.basename(image_paths[i]))
                
                # Save visualization if requested
                if save_dir and labels[i].item() == 1:  # Only save anomalous samples
                    save_visualization(
                        images[i], 
                        anomaly_map, 
                        masks[i] if masks[i] is not None else torch.zeros((224, 224)),
                        os.path.join(save_dir, os.path.basename(image_paths[i])),
                        category
                    )
    
    return {
        'image_scores': np.array(image_scores),
        'pixel_scores': pixel_scores,
        'true_labels': np.array(true_labels),
        'true_masks': true_masks,
        'file_names': file_names
    }
```

```python
def save_visualization(image, anomaly_map, mask, save_path, category=None):
    """Save visualization of original image, anomaly map, and ground truth."""
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    
    # Original image
    img = image.cpu().permute(1, 2, 0).numpy()
    # Denormalize
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    img = img * std + mean
    img = np.clip(img, 0, 1)
    axes[0].imshow(img)
    axes[0].set_title('Original')
    axes[0].axis('off')
    
    # Anomaly map
    axes[1].imshow(anomaly_map, cmap='plasma')
    axes[1].set_title('Anomaly Map')
    axes[1].axis('off')
    
    # Ground truth mask
    mask_np = mask.cpu().numpy()
    axes[2].imshow(mask_np, cmap='gray')
    axes[2].set_title('Ground Truth')
    axes[2].axis('off')
    
    if category:
        plt.suptitle(f"Category: {category}")
    
    plt.tight_layout()
    plt.savefig(save_path)
    plt.close()
```

```python
def evaluate_model(results):
    """Calculate evaluation metrics with optimal thresholds for F1 scores."""
    image_scores = results['image_scores']
    pixel_scores = results['pixel_scores']
    true_labels = results['true_labels']
    true_masks = results['true_masks']
    
    # Calculate image-level metrics
    image_auroc = roc_auc_score(true_labels, image_scores)
    
    # Find optimal image-level threshold for F1
    image_thresholds = np.linspace(np.min(image_scores), np.max(image_scores), 100)
    best_image_f1 = 0
    best_image_threshold = 0
    
    for threshold in image_thresholds:
        pred_labels = (image_scores > threshold).astype(int)
        if np.sum(pred_labels) > 0 and np.sum(true_labels) > 0:  # Avoid empty predictions
            f1 = f1_score(true_labels, pred_labels)
            if f1 > best_image_f1:
                best_image_f1 = f1
                best_image_threshold = threshold
    
    # Calculate image-level precision/recall at best threshold
    if best_image_f1 > 0:
        pred_labels = (image_scores > best_image_threshold).astype(int)
        image_precision = precision_score(true_labels, pred_labels)
        image_recall = recall_score(true_labels, pred_labels)
    else:
        image_precision = 0
        image_recall = 0
    
    # Flatten all pixel scores and masks
    all_pixels_scores = []
    all_pixels_masks = []
    
    for i, mask in enumerate(true_masks):
        # Only use anomalous samples for pixel-level evaluation
        if true_labels[i] == 1:
            score = pixel_scores[i].flatten()
            # Ensure mask is binary (0 or 1)
            mask_flat = (mask > 0).astype(np.int32).flatten()
            
            all_pixels_scores.append(score)
            all_pixels_masks.append(mask_flat)
    
    if all_pixels_scores and np.sum(all_pixels_masks[0]) > 0:
        all_pixels_scores = np.concatenate(all_pixels_scores)
        all_pixels_masks = np.concatenate(all_pixels_masks)
        
        # Calculate pixel-level AUROC
        try:
            pixel_auroc = roc_auc_score(all_pixels_masks, all_pixels_scores)
            
            # Calculate best F1 score
            precision, recall, thresholds = precision_recall_curve(all_pixels_masks, all_pixels_scores)
            f1_scores = 2 * precision * recall / (precision + recall + 1e-8)
            best_threshold = thresholds[np.argmax(f1_scores)]
            best_f1 = np.max(f1_scores)
            
            result_dict = {
                'image_auroc': image_auroc,
                'image_f1_score': best_image_f1,
                'image_threshold': best_image_threshold,
                'image_precision': image_precision,
                'image_recall': image_recall,
                'pixel_auroc': pixel_auroc,
                'pixel_f1_score': best_f1,
                'pixel_threshold': best_threshold
            }
        except Exception as e:
            print(f"Error calculating pixel-level metrics: {e}")
            result_dict = {
                'image_auroc': image_auroc,
                'image_f1_score': best_image_f1,
                'image_threshold': best_image_threshold,
                'image_precision': image_precision,
                'image_recall': image_recall
            }
    else:
        result_dict = {
            'image_auroc': image_auroc,
            'image_f1_score': best_image_f1,
            'image_threshold': best_image_threshold,
            'image_precision': image_precision,
            'image_recall': image_recall
        }
    
    return result_dict
```

```python
def plot_score_distributions(category, results, save_dir=None):
    """Plot score distributions with optimal thresholds."""
    image_scores = results['image_scores']
    true_labels = results['true_labels']
    
    metrics = evaluate_model(results)
    best_threshold = metrics.get('image_threshold', 0)
    
    # Separate scores for normal and anomalous samples
    normal_scores = image_scores[true_labels == 0]
    anomaly_scores = image_scores[true_labels == 1]
    
    plt.figure(figsize=(12, 6))
    
    # Plot histograms
    plt.hist(normal_scores, bins=30, alpha=0.5, label='Normal', density=True)
    plt.hist(anomaly_scores, bins=30, alpha=0.5, label='Anomaly', density=True)
    
    # Plot optimal threshold
    plt.axvline(x=best_threshold, color='r', linestyle='--', 
                label=f'Threshold: {best_threshold:.4f} (F1: {metrics.get("image_f1_score", 0):.4f})')
    
    plt.title(f'Score Distribution for {category}')
    plt.xlabel('Anomaly Score')
    plt.ylabel('Density')
    plt.legend()
    
    if save_dir:
        plt.savefig(os.path.join(save_dir, f'{category}_score_distribution.png'))
    
    plt.show()
    return metrics
```

```python
def comprehensive_benchmark(feature_extractor, adapter, discriminator, 
                           data_path, category, device, 
                           batch_size=4, num_iterations=50):
    """
    Comprehensive benchmark that includes the complete inference pipeline:
    - Data loading
    - Preprocessing
    - Model inference
    - Post-processing (generating anomaly maps)
    """
    # Set models to evaluation mode
    feature_extractor.eval()
    adapter.eval()
    discriminator.eval()
    
    # Create a test dataloader with real data
    test_loader = get_test_dataloader(data_path, category, batch_size=batch_size)
    
    # Warm-up run - process a few batches to warm up the GPU
    print("Warming up...")
    with torch.no_grad():
        for images, _, _, _ in test_loader:
            if len(images) != batch_size:  # Skip incomplete batches
                continue
                
            images = images.to(device)
            feats = feature_extractor(images)
            combined_feats = pool_and_combine_features(feats)
            embedding = flatten_features(combined_feats)
            adapted = adapter(embedding)
            scores = discriminator(adapted)
            
            # Only process one batch for warm-up
            break
            
    # Measure time for the complete pipeline
    print(f"Benchmarking with batch size {batch_size}...")
    torch.cuda.synchronize()
    start_time = time.time()
    
    # Count total processed images
    total_images = 0
    
    with torch.no_grad():
        # Process data for a fixed number of iterations
        iterations = 0
        while iterations < num_iterations:
            # Loop through the dataloader
            for images, labels, masks, image_paths in test_loader:
                if len(images) != batch_size:  # Skip incomplete batches for consistent timing
                    continue
                    
                # Transfer to device
                images = images.to(device)
                
                # Model inference
                feats = feature_extractor(images)
                combined_feats = pool_and_combine_features(feats)
                embedding = flatten_features(combined_feats)
                adapted = adapter(embedding)
                scores = discriminator(adapted)
                
                # Post-processing (generate anomaly maps)
                scores_map = -scores.view(batch_size, combined_feats.shape[2], combined_feats.shape[3])
                
                # For each image, resize and apply Gaussian filter (as in the paper)
                for i in range(batch_size):
                    anomaly_map = scores_map[i].cpu().numpy()
                    resized_map = cv2.resize(anomaly_map, (224, 224))
                    filtered_map = gaussian_filter(resized_map, sigma=4)
                    
                    # You could calculate image-level score here if needed
                    img_score = np.max(filtered_map)
                
                total_images += batch_size
                iterations += 1
                
                if iterations >= num_iterations:
                    break
    
    torch.cuda.synchronize()
    end_time = time.time()
    
    # Calculate elapsed time and FPS
    elapsed_time = end_time - start_time
    fps = total_images / elapsed_time
    
    print(f"Processed {total_images} images in {elapsed_time:.2f} seconds")
    print(f"Complete pipeline FPS: {fps:.2f} images/sec")
    
    return fps
```
